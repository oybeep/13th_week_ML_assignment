```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from google.colab import drive
drive.mount('/content/drive')
```

```python
df=pd.read_csv('/content/drive/MyDrive/drug200.csv')
```

```python
df['Sex']=df['Sex'].replace({'M':1,'F':2})
df['BP']=df['BP'].replace({'LOW':1,'NORMAL':0,'HIGH':2})
df['Cholesterol']=df['Cholesterol'].replace({'NORMAL':1,'HIGH':2})
df['Drug']=df['Drug'].replace({'drugA':0,'drugB':1,'drugC':2,'drugX':3,'drugY':4})

X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
clf = DecisionTreeClassifier()
clf.fit(X, y)

plt.figure(figsize=(20, 10))
plot_tree(clf, feature_names=['Age','Sex','Cholesterol','Na_to_K','Drug'],class_names=['drug A', 'drug B', 'drug C', 'drug X', 'drug Y',],filled=True)
plt.title("Decision Tree for Drug Prediction")
plt.show()
```
